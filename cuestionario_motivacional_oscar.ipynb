```python
# Convert to datetime the column we use to determine the last day of a session execution
df_not_cleaned['updated_at_ex_ex'] = pd.to_datetime(df_not_cleaned['updated_at_ex_ex'])

# get the id_users based on the maximum date in which the users executed a session
idx = df_not_cleaned.groupby('id_users')['updated_at_ex_ex'].idxmax()

df_last_section_execution = df_not_cleaned.loc[idx] # We replace the option replaced of first by the maximum date of a session, since she for her experiments did not care about the maximum session execution day

df_last_section_execution = df_last_section_execution.loc[:, ['id_users', 'created_at_users', 'updated_at_ex_ex']]
df_last_section_execution['id_users'] = df_last_section_execution['id_users'].astype('int64') 
```

```python
df_motivational_filtered = df_motivational_raw.merge(df_last_section_execution, how = 'inner', left_on = 'id', right_on = 'id_users')
df_motivational_filtered['created_at_users'] =pd.to_datetime(df_motivational_filtered['created_at_users'])
df_motivational_filtered = df_motivational_filtered[df_motivational_filtered['created_at_users'] > '2021-01-01']
df_motivational_filtered['days_since_enrolment'] = df_motivational_filtered['updated_at_ex_ex'] - df_motivational_filtered['created_at_users']
```
