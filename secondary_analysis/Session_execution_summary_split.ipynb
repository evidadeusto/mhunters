```python
def split(x):
    parts = []
    bracket_level = 0
    current = []
    for c in (x + ","):
        if c == "," and bracket_level == 0:
            parts.append("".join(current))
            current = []
        else:
            if c == "{":
                bracket_level += 1
            elif c == "}":
                bracket_level -= 1
            current.append(c)
    return parts
```

```python
def conversion_from_jsonb(df, var):
    if var in ['reps_per_exercise', 'mins_per_exercise',
              'reps_per_min_per_exercise',  'reps_set_per_block', 
              'time_set_per_block', 'average_reps_min_set_per_block']:
        
        data = pd.DataFrame(df[var])
        data[var] = data[var].astype("string")
        data = data.dropna()
        
        x = []

        for i, v in data[var].items():
            #print(data[var][i])
            data[var][i] = data[var][i].replace("{", "")
            data[var][i] = data[var][i].replace("}", "")
            data[var][i] = data[var][i].replace(":", ",")
            #jdata[var][i] = data[var][i].str.replace('"', '')

            #data[var][i] = data[var][i].split(",")
            data[var][i] = data[var][i].replace("(shoulder, hip, knee, ankle)", "")

            x.append(data[var][i].split(","))
        
        data2 = pd.DataFrame(x)
        data2 = data2.replace('"', '', regex=True)
        data2 = data2.replace(" null", np.NaN)
        
        data2.fillna("",inplace=True)
        
        
        if var == 'reps_per_exercise':
            
            data2 = data2.loc[:, ~(data2 == ' reps').any()]
            data2 = data2.loc[:, ~(data2 == ' name_en').any()]
            data2 = data2.loc[:, ~(data2 == ' name_es').any()]
            
            maxcol = int(len(data2.columns)/4)
            
            basename1 = "no"
            no = ["{}_{}".format(basename1, i) for i in range(1,maxcol+1)]

            basename2 = 'reps'
            reps = ["{}_{}".format(basename2, i) for i in range(1,maxcol+1)]

            basename3 = 'name_en'
            name_en = ["{}_{}".format(basename3, i) for i in range(1,maxcol+1)]

            basename4 = 'name_es'
            name_es = ["{}_{}".format(basename4, i) for i in range(1,maxcol+1)]
   
            y = []

            for j in range(0, maxcol):
                y.append(no[j])
                y.append(reps[j])
                y.append(name_en[j])
                y.append(name_es[j])

            data2.columns = y
        
            for header in data2.columns:
                if "reps" in header:
                    #print(maybe4[header].head())
                    data2[header] = pd.to_numeric(data2[header], errors='coerce')
                    #data2[header] = data2[header].astype('Int64')

                if "no_" in header:
                    #maybe4[header].head()
                    data2[header] = pd.to_numeric(data2[header], errors='coerce')
                    #data2[header] = data2[header].astype('Int64')
                
                """
                if "name_" in header:
                    data2[header] = data2[header].fillna(None)
                """

        
        if var == 'mins_per_exercise':
            
            data2 = data2.loc[:, ~(data2 == ' execution_time').any()]
            data2 = data2.loc[:, ~(data2 == ' name_en').any()]
            data2 = data2.loc[:, ~(data2 == ' name_es').any()]
            
            maxcol = int(len(data2.columns)/4)
            
            basename1 = "no"
            no = ["{}_{}".format(basename1, i) for i in range(1,maxcol+1)]

            basename2 = 'name_en'
            name_en = ["{}_{}".format(basename2, i) for i in range(1,maxcol+1)]

            basename3 = 'name_es'
            name_es = ["{}_{}".format(basename3, i) for i in range(1,maxcol+1)]
            
            basename4 = 'execution_time'
            execution_time = ["{}_{}".format(basename4, i) for i in range(1,maxcol+1)]
            
            y = []

            for j in range(0, maxcol):
                y.append(no[j])
                y.append(name_en[j])
                y.append(name_es[j])
                y.append(execution_time[j])

            data2.columns = y
        
            for header in data2.columns:
                if "execution_time" in header:
                    #print(maybe4[header].head())
                    data2[header] = pd.to_numeric(data2[header], errors='coerce')
                    data2[header] = data2[header].astype('Int64')

                if "no_" in header:
                    #maybe4[header].head()
                    data2[header] = pd.to_numeric(data2[header], errors='coerce')
                    data2[header] = data2[header].astype('Int64')
        
        if var == 'reps_per_min_per_exercise':
            
            data2 = data2.loc[:, ~(data2 == ' value').any()]
            data2 = data2.loc[:, ~(data2 == ' name_en').any()]
            data2 = data2.loc[:, ~(data2 == ' name_es').any()]
            
            maxcol = int(len(data2.columns)/4)
            
            basename1 = "no"
            no = ["{}_{}".format(basename1, i) for i in range(1,maxcol+1)]

            basename2 = 'value'
            val = ["{}_{}".format(basename2, i) for i in range(1,maxcol+1)]

            basename3 = 'name_en'
            name_en = ["{}_{}".format(basename3, i) for i in range(1,maxcol+1)]

            basename4 = 'name_es'
            name_es = ["{}_{}".format(basename4, i) for i in range(1,maxcol+1)]
            
            y = []

            for j in range(0, maxcol):
                y.append(no[j])
                y.append(val[j])
                y.append(name_en[j])
                y.append(name_es[j])
                

            data2.columns = y
        
            for header in data2.columns:
                if "value" in header:
                    #print(maybe4[header].head())
                    data2[header] = pd.to_numeric(data2[header], errors='coerce')
                    data2[header] = data2[header].astype('Int64')

                if "no_" in header:
                    #maybe4[header].head()
                    data2[header] = pd.to_numeric(data2[header], errors='coerce')
                    data2[header] = data2[header].astype('Int64')
                    
'''
        if var == 'reps_set_per_block':
            
            data2 = data2.iloc[1:]
            
            data2 = data2.loc[:, ~(data2 == ' reps').any()]
            
            maxcol = int(len(data2.columns)/2)
            
            no = ['no']
            
            basename2 = 'no1'
            no1 = ["{}_{}".format(basename2, i) for i in range(1,maxcol+1)]

            basename3 = 'value'
            val = ["{}_{}".format(basename3, i) for i in range(1,maxcol+1)]
            
            y = []
            y.append(no[0])

            for j in range(0, maxcol):
                y.append(no1[j])
                y.append(val[j])
                
            data2.columns = y
        
            for header in data2.columns:
                if "value" in header:
                    #print(maybe4[header].head())
                    data2[header] = pd.to_numeric(data2[header], errors='coerce')
                    data2[header] = data2[header].astype('Int64')

                if "no_" in header:
                    #maybe4[header].head()
                    data2[header] = pd.to_numeric(data2[header], errors='coerce')
                    data2[header] = data2[header].astype('Int64')
                    
                if "no" in header:
                    #maybe4[header].head()
                    data2[header] = pd.to_numeric(data2[header], errors='coerce')
                    data2[header] = data2[header].astype('Int64')
                    
                   

        if var == 'time_set_per_block':
            
            data2 = data2.loc[:, ~(data2 == ' execution_time').any()]
            
            maxcol = int(len(data2.columns)/2)
            
            no = ['no']
            
            basename2 = 'no1'
            no1 = ["{}_{}".format(basename2, i) for i in range(1,maxcol+1)]

            basename3 = 'execution_time'
            execution_time = ["{}_{}".format(basename3, i) for i in range(1,maxcol+1)]
            
            y = []
            y.append(no[0])

            for j in range(0, maxcol):
                y.append(no1[j])
                y.append(execution_time[j])
                
            data2.columns = y
        
            for header in data2.columns:
                if "execution_time" in header:
                    #print(maybe4[header].head())
                    data2[header] = pd.to_numeric(data2[header], errors='coerce')
                    data2[header] = data2[header].astype('Int64')

                if "no_" in header:
                    #maybe4[header].head()
                    data2[header] = pd.to_numeric(data2[header], errors='coerce')
                    data2[header] = data2[header].astype('Int64')
                    
                if "no" in header:
                    #maybe4[header].head()
                    data2[header] = pd.to_numeric(data2[header], errors='coerce')
                    data2[header] = data2[header].astype('Int64') 

'''

        if var == 'average_reps_min_set_per_block':
            
            # data2 = data2.loc[:, ~(data2 == ' execution_time').any()]
            
            maxcol = int(len(data2.columns)/2)
            
            no = ['no']
            
            basename2 = 'no1'
            no1 = ["{}_{}".format(basename2, i) for i in range(1,maxcol+1)]

            basename3 = 'average'
            average = ["{}_{}".format(basename3, i) for i in range(1,maxcol+1)]
            
            y = []
            y.append(no[0])

            for j in range(0, maxcol):
                y.append(no1[j])
                y.append(average[j])
                
            y.pop()
                
            data2.columns = y
            
            for header in data2.columns:
                if "average" in header:
                    #print(maybe4[header].head())
                    data2[header] = pd.to_numeric(data2[header], errors='coerce')

                if "no1_" in header:
                    #maybe4[header].head()
                    data2[header] = pd.to_numeric(data2[header], errors='coerce')
                    #data2[header] = data2[header].astype('Int64')
                    
                if "no" in header:
                    #maybe4[header].head()
                    data2[header] = pd.to_numeric(data2[header], errors='coerce')
                    #data2[header] = data2[header].astype('Int64') 
                    
        
        
        #data2.fillna("",inplace=True)
        data2.fillna(np.NaN, inplace = True)
        data2 = data2.replace("", np.NaN)
        #data2 = data2.where(~pd.isna(data2), np.nan)
        data2 = data2.replace(pd.NaT, np.NaN)
            
        return(data2)
         
    else: print("Sorry! Wrong variable chosen!")
```
